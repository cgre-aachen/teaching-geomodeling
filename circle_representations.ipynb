## Circle Representations

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('default')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
circ = plt.Circle((0, 0), radius=1, edgecolor='b', facecolor='None')
ax.add_patch(circ)
plt.show()
```

```python
x = np.arange(-1.2,1.2,0.01)
y = np.arange(-1.2,1.2,0.01)
XX,YY = np.meshgrid(x, y)

implicit_circle = XX**2 + YY**2
```

```python
plt.imshow(implicit_circle, extent=(-1.2,1.2,-1.2,1.2))
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1, 1, 1)
ax.axhline(0, color='k', lw=2)
ax.axvline(0, color='k', lw=2)
t = np.arange(0,2*np.pi,0.01)
ax.plot(np.cos(t), np.sin(t), linewidth=3)

ax.imshow(implicit_circle, extent=(-1.2,1.2,-1.2,1.2))

ax.axis('equal');
ax.set_xlim([-1.2,1.2])
ax.set_ylim([-1.2,1.2])

ax.grid()


```

```python
from mpl_toolkits import mplot3d
```

```python

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(XX, YY, implicit_circle, cmap='viridis', edgecolor='none', alpha=0.5)
ax.plot_surface(XX, YY, np.ones_like(XX), cmap='viridis', edgecolor='none', alpha=0.2)
ax.plot(np.cos(t), np.sin(t), 1, linewidth=3, c='black')



ax.set_title('Implicit circle')
plt.show()
```
